```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# create the dataset and display iris dataframe
iris = pd.read_csv(r"C:\Users\udayk\OneDrive\B_A (analyst)\Logistic_Regression\Iris.csv")
iris.head()
```

```python
iris['Species']
```

```python
iris.columns
```

```python
# split the data into features(X) and labels(y) 
X=iris.drop(columns=['Id','Species'])
y=iris['Species']
```

```python
# split the data into training set and test set
X_train,X_test,y_train, y_test= train_test_split(X,y,test_size=0.2, random_state=40)
```

```python
# Standardize the features
scaler= StandardScaler()
x_train_scaled= scaler.fit_transform(X_train)
x_test_scaled=scaler.transform(X_test)
```

```python
# create the ml model
model=LogisticRegression()
```

```python
# train the model
model.fit(x_train_scaled,y_train)
```

```python
# Evalute the model on the Testing data
y_pred= model.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy =accuracy_score(y_test, y_pred)
print("Accuracy",accuracy)
```

```python

print("Classification report",classification_report(y_test,y_pred))
```

```python
model.score(x_test_scaled,y_test)
```

```python
prediction =model.predict([[4.2,3.2,1.2,0.2]])
print(prediction)
```
